```python
# 1. Import all the AI libraries we'll need
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
import numpy as np

print(f"Using TensorFlow version: {tf.__version__}")

# 2. Define our main variables
IMAGE_SIZE = (128, 128)  # We will resize all images to 128x128 pixels.
BATCH_SIZE = 32          # We will feed the AI 32 images at a time.

# THIS IS THE LINE WE FIXED!
# We are now pointing to the correct folder with the space and capital 'G'.
DATA_DIR = './data/Garbage classification'

# 3. Create the data "pipelines"
# 'validation_split=0.2' means 20% of data is for testing.
# 'subset="training"' means this is the 80% for training.
train_dataset = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR,
    validation_split=0.2,
    subset="training",
    seed=123, # 'seed' just makes sure the split is reproducible
    image_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE
)

# 'subset="validation"' means this is the 20% for testing.
validation_dataset = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE
)

# 4. Get the class names
class_names = train_dataset.class_names
print("\n--- Found the following classes: ---")
print(class_names)

# 5. Let's look at some of the data!
print("\n--- Displaying a sample of 9 training images ---")
plt.figure(figsize=(10, 10))
for images, labels in train_dataset.take(1):
  # .take(1) grabs one batch (32 images)
  for i in range(9):
    # We will show the first 9 images from that batch
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(class_names[labels[i]])
    plt.axis("off")

plt.show()
```

```python
# --- Part 1: Data Augmentation ---
# To make our model "amazing" and not "mediocre," we will add a layer
# that randomly flips, rotates, and zooms the images.
# This prevents the AI from just "memorizing" the exact pictures.
data_augmentation = keras.Sequential(
    [
        layers.RandomFlip("horizontal"),
        layers.RandomRotation(0.1),
        layers.RandomZoom(0.1),
    ]
)

# --- Part 2: Build the Model ---
# We are building a lightweight model, perfect for Edge AI.
model = keras.Sequential([
    # Input layer: Rescale all pixel values from [0, 255] to [0, 1]
    layers.Rescaling(1./255, input_shape=(IMAGE_SIZE[0], IMAGE_SIZE[1], 3)),

    # Apply our random augmentations
    data_augmentation,

    # 1st "convolutional block"
    # It learns 16 different simple features (edges, curves)
    layers.Conv2D(16, (3, 3), padding='same', activation='relu'),
    layers.MaxPooling2D(),

    # 2nd "convolutional block"
    # It learns 32 more complex features (combinations of edges)
    layers.Conv2D(32, (3, 3), padding='same', activation='relu'),
    layers.MaxPooling2D(),

    # 3rd "convolutional block"
    # It learns 64 even more complex features (shapes, textures)
    layers.Conv2D(64, (3, 3), padding='same', activation='relu'),
    layers.MaxPooling2D(),

    # "Flatten" the 3D feature data into a 1D list
    layers.Flatten(),

    # A "Dense" (or fully connected) layer to analyze the features
    layers.Dense(128, activation='relu'),

    # The "Output" layer. It has 6 nodes (one for each class).
    # 'softmax' converts the output into a probability (e.g., 80% plastic)
    layers.Dense(len(class_names), activation='softmax')
])

# --- Part 3: Compile the Model ---
model.compile(
    optimizer='adam', # 'adam' is a popular, efficient optimizer
    loss=tf.keras.losses.SparseCategoricalCrossentropy(), # Good for multi-class classification
    metrics=['accuracy'] # We want to track the accuracy
)

# Let's look at the "blueprint" of our model
print("--- MODEL BLUEPRINT ---")
model.summary()

# --- Part 4: Train the Model ---
EPOCHS = 15 # We will loop through the *entire* dataset 15 times.

print("\n--- STARTING TRAINING ---")
# This is the command that starts the training!
history = model.fit(
    train_dataset,
    validation_data=validation_dataset,
    epochs=EPOCHS
)
print("--- TRAINING COMPLETE ---")

# --- Part 5: Show the Results ---
# We will plot the "accuracy" and "loss" over time.
# This is the #1 way to see if our AI learned correctly.

# 'history' contains all the data from our training run
acc = history.history['accuracy']
val_acc = history.history['validation_accuracy']
loss = history.history['loss']
val_loss = history.history['validation_loss']

epochs_range = range(EPOCHS)

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')

plt.show()
```

```python
# --- Part 1: Data Augmentation ---
# (This part is the same)
data_augmentation = keras.Sequential(
    [
        layers.RandomFlip("horizontal"),
        layers.RandomRotation(0.1),
        layers.RandomZoom(0.1),
    ]
)

# --- Part 2: Build the Model ---
model = keras.Sequential([
    layers.Rescaling(1./255, input_shape=(IMAGE_SIZE[0], IMAGE_SIZE[1], 3)),
    data_augmentation,
    layers.Conv2D(16, (3, 3), padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(32, (3, 3), padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64, (3, 3), padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),

    # --- THIS IS THE CRITICAL FIX ---
    # Instead of relying on a variable, we are "hard-coding" the
    # 6 classes. This is much safer and will fix the 0.0 accuracy.
    layers.Dense(6, activation='softmax')
])

# --- Part 3: Compile the Model ---
model.compile(
    optimizer='adam',
    # We use 'SparseCategoricalCrossentropy' because our labels are numbers (0, 1, 2...)
    # not "one-hot" vectors.
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    metrics=['accuracy']
)

print("--- MODEL BLUEPRINT (FIXED) ---")
model.summary()

# --- Part 4: Train the Model ---
EPOCHS = 15

print("\n--- STARTING TRAINING ---")
history = model.fit(
    train_dataset,
    validation_data=validation_dataset,
    epochs=EPOCHS
)
print("--- TRAINING COMPLETE ---")

# --- Part 5: Show the Results ---
# 'history' contains all the data from our training run
acc = history.history['accuracy']
loss = history.history['loss']

# --- THIS IS THE TYPO FIX ---
# The log saves 'val_accuracy', not 'validation_accuracy'.
# We are fixing that typo here.
val_acc = history.history['val_accuracy']
val_loss = history.history['val_loss']

epochs_range = range(EPOCHS)

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')

plt.show()
```

```python
import tensorflow as tf

# 1. Create a converter object from our trained model
converter = tf.lite.TFLiteConverter.from_keras_model(model)

# 2. Apply standard optimizations to make it smaller and faster
converter.optimizations = [tf.lite.Optimize.DEFAULT]

# 3. Convert the model!
tflite_model = converter.convert()

# 4. Save the converted model to a file
with open('model.tflite', 'wb') as f:
  f.write(tflite_model)

print("--- TFLite Model Saved as 'model.tflite' ---")

# 5. Let's check the new file size!
# The '-lh' flag means 'list in human-readable' format (e.g., MB or KB)
!ls -lh model.tflite
```

```python
# 1. Load the validation dataset again so we can grab one test image
validation_dataset = tf.keras.utils.image_dataset_from_directory(
    DATA_DIR,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=IMAGE_SIZE,
    batch_size=1  # We only want to grab one image
)

# 2. Get the class names
class_names = validation_dataset.class_names

# 3. Grab one image and its label from the validation set
for image, label in validation_dataset.take(1):
  test_image = image
  true_label = label.numpy()[0]

print(f"--- Preparing to test one image... ---")
print(f"True Class: {class_names[true_label]}")

# 4. Load the TFLite model into the "Interpreter"
interpreter = tf.lite.Interpreter(model_path='model.tflite')
interpreter.allocate_tensors()

# 5. Get the input and output details from the model
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

# 6. Set the test image as the model's input
interpreter.set_tensor(input_details[0]['index'], test_image)

# 7. Run the prediction!
interpreter.invoke()

# 8. Get the prediction results
output_data = interpreter.get_tensor(output_details[0]['index'])
predicted_label_index = np.argmax(output_data)

print(f"Model Prediction: {class_names[predicted_label_index]}")

# 9. Show the result
if class_names[predicted_label_index] == class_names[true_label]:
  print("\n*** Success! The model was correct. ***")
else:
  print("\n*** Failure. The model was incorrect. ***")

# 10. Show the image we tested
plt.imshow(test_image[0].numpy().astype("uint8"))
plt.title(f"True: {class_names[true_label]} | Predicted: {class_names[predicted_label_index]}")
plt.axis("off")
plt.show()
```
